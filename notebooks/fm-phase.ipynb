```python
import numpy as np
from lineage.LineageTree import LineageTree
from lineage.states.StateDistPhase import StateDistribution2

pi = np.array([0.6, 0.4], dtype="float");
T = np.array([[0.75, 0.25],
              [0.15, 0.85]], dtype="float");

bern_p0 = 0.99
gamma_a0g1 = 7
gamma_scale0g1 = 7
gamma_a0g2 = 10
gamma_scale0g2 = 10
# State 1 parameters "Susceptible"
bern_p1 = 0.88
gamma_a1g1 = 7
gamma_scale1g1 = 1
gamma_a1g2 = 10
gamma_scale1g2 = 3

state_obj0 = StateDistribution2(bern_p0, gamma_a0g1, gamma_scale0g1, gamma_a0g2, gamma_scale0g2)
state_obj1 = StateDistribution2(bern_p1, gamma_a1g1, gamma_scale1g1, gamma_a1g2, gamma_scale1g2)

E = [state_obj0, state_obj1]
```

```python
lineage1 = LineageTree(pi, T, E, desired_num_cells=2**8 - 1)
# These are the minimal arguments required to instantiate lineages
print(lineage1)
print("\n")
```

```python
from lineage.Analyze import Analyze

X = [lineage1] # population just contains one lineage
tHMMobj, pred_states_by_lineage, LL = Analyze(X, 2);
```

```python
print(tHMMobj.estimate.pi)
```

```python
print(tHMMobj.estimate.T)
```
