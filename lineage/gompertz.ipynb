```python
import scipy.stats as sp
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5)
c_list = np.linspace(0.1, 3, num=8)
for c in c_list:
    plt.plot(x, sp.gompertz.pdf(x, c), alpha=0.6, label=str(c))
    
plt.legend(loc='upper right', bbox_to_anchor=(1.5, 1.05))
```

```python
x = np.linspace(0, 5)
scale_list = np.linspace(0.5, 3, num=8)
for i in scale_list:
    plt.plot(x, sp.gompertz.pdf(x, c=0.75, scale=i), alpha=0.6, label=str(i))
    
plt.legend(loc='upper right', bbox_to_anchor=(1.5, 1.05))
```

```python
cGom = 0.75
tau = np.zeros((500, 8))

for i in range(8):
    for n in range(500):
        tau[n, i] = sp.gompertz.rvs(cGom, scale=scale_list[i])
```

```python
for n in range(8):
    plt.hist(tau[:, n], label=str(scale_list[n]))
    plt.xlim(0,6)
    plt.ylim(0,100)
    plt.xlabel("tau")
    plt.ylabel("#")
    plt.legend()
    plt.show()
```

```python
scaleGom = 1.25
tau = np.zeros((500, 8))

for i in range(8):
    for n in range(500):
        tau[n, i] = sp.gompertz.rvs(c=c_list[i], scale=scaleGom)
        
for n in range(8):
    plt.hist(tau[:, n], label=str(c_list[n]))
    plt.xlim(0,5)
    plt.ylim(0,170)
    plt.xlabel("tau")
    plt.ylabel("#")
    plt.legend()
    plt.show()
```
