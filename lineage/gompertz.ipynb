```python
import scipy.stats as sp
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5)
c_list = np.linspace(0.1, 3, num=8)
for c in c_list:
    plt.plot(x, sp.gompertz.pdf(x, c), alpha=0.6, label=str(c))
    
plt.legend(loc='upper right', bbox_to_anchor=(1.5, 1.05))
```

```python
x = np.linspace(0, 5)
scale_list = np.linspace(0.1, 3, num=8)
for i in scale_list:
    plt.plot(x, sp.gompertz.pdf(x, c=0.75, scale=i), alpha=0.6, label=str(i))
    
plt.legend(loc='upper right', bbox_to_anchor=(1.5, 1.05))
```
